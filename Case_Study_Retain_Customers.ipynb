9. How many customers have churned?

```python
df1 = df[df['Churn']=='Yes']
churned = df1['Churn'].value_counts()
churned
```

10. What is the churn rate in the dataset?


```python
ttl = len(df['Churn'])
churn_rate = round((churned/ttl)*100,2)
churn_rate
```

11. How does churn vary across different genders?


```python
chrn = df[df['Churn']=='Yes']
gen_chrn = chrn.groupby('gender')[['Churn']].count()
gen_chrn
```

12. Are there any differences in churn based on the customer's age?

```python
age_chrn = chrn.groupby('SeniorCitizen')['Churn'].count().reset_index()
age_chrn.rename(columns={'SeniorCitizen':'Citizen'},inplace=True)
age_chrn['Citizen'] = ['Non Senior Citizen','Senior Citizen']
age_chrn
```
